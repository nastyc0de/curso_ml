# Plots y visualizaciones de los datos

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('./datasets/customer-churn-model/Customer Churn Model.txt')
data.shape
```

```python
% matplotlib inline
```

### Scatter Plot

```python
data.plot(kind='scatter', x='Night Mins', y='Night Charge')
```

```python
data.plot(kind='scatter', x='Day Mins', y='Day Charge')
```

```python
figure, axe = plt.subplots(2, 2, sharex=True, sharey=True)
data.plot(kind='scatter', x='Day Mins', y='Day Charge', ax=axe[0][0])
data.plot(kind='scatter', x='Night Mins', y='Night Charge', ax=axe[0][1])
data.plot(kind='scatter', x='Day Calls', y='Day Charge', ax=axe[1][0])
data.plot(kind='scatter', x='Night Calls', y='Night Charge', ax=axe[1][1])
```

### Histogramas de frecuencia

```python
plt.hist(data['Day Calls'], bins=round(1 + np.log2(3333)))
plt.xlabel('Numero de llamadas al dia')
plt.ylabel('Frecuencia')
plt.title('Histograma de numeros de llamadas al dia')
```

### Boxplot, diagrama de caja y bigotes

```python
plt.boxplot(data['Day Calls'])
plt.ylabel('Nro llamadas diarias')
plt.title('Boxplot de llamadas diarias')
```
